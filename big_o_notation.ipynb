```python
import matplotlib.pyplot as plt
import math as m
```

```python
x = range(1, 4)
y1 = [0 * n + 1 for n in x]
y2 = [m.log(n) for n in x]
y3 = [n for n in x]
y4 = [n * m.log(n) for n in x]
y5 = [n ** 2 for n in x]
y6 = [2 ** n for n in x]
y7 = [m.factorial(n) for n in x]
y8 = [n ** n for n in x]
plt.figure(figsize=(15, 10))
plt.scatter(x, y1)
plt.plot(x, y1, ls= ':', label = '$\O(1)$')
plt.scatter(x, y2)
plt.plot(x, y2, ls= 'solid', label= '$\O(log n)$')
plt.scatter(x, y3)
plt.plot(x, y3, ls= '--', label= '$\O(n)$')
plt.scatter(x, y4)
plt.plot(x, y4, ls= 'dashdot', label= "$\O(n log n)$")
plt.scatter(x, y5)
plt.plot(x, y5, ls= ':', label = '$\O(n^2)$')
plt.scatter(x, y6)
plt.plot(x, y6, ls= 'solid', label = "$\O(2^n)$")
plt.scatter(x, y7)
plt.plot(x, y7, ls= '--', label = "$\O(n!)$")
plt.scatter(x, y8)
plt.plot(x, y8, ls= 'dashdot', label = "$\O(n^n)$")
plt.title('Big O Notation')
plt.xlabel('heretoo')
plt.ylabel('here')
plt.legend(loc='upper left')
```

```python
plt.show()
```
