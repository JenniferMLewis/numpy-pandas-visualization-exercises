```python
import matplotlib.pyplot as plt
import math as m
from random import randint
```

```python
# list comprehension: [n for n in x]
```

```python
x = range(-20,20)
y = [n**2 - n +2 for n in x]
# arrowprops and arrowstyle are amazingly nice ;;
plt.plot(x,y)
plt.annotate('origin', xy=(0,0), xytext = (0,100), arrowprops= dict (arrowstyle = '->'))
plt.show()
```

```python
plt.figure(figsize=(10,7))
x1 = range(1, 10)
y1 = [m.sqrt(n) for n in x1]
plt.subplot(221)
plt.plot(x1, y1)
plt.title('square root of x')

x2 = range(-8, 8)
y2 = [n ** 3 for n in x2]
plt.subplot(222)
plt.plot(x2, y2)
plt.title('x^3')

x3 = range(0, 10)
y3 = [2 ** n for n in x3]
plt.subplot(223)
plt.plot(x3, y3)
plt.title('2^x')

x4 = range(0, 8)
y4 = [1/(n + 1) for n in x4]
plt.subplot(224)
plt.plot(x4, y4)
plt.title('1/(x+1)')
#Book says 1 / x + 1, not sure why review was 1 / x - 1


plt.suptitle("Four Functions")
plt.show()
```

```python
plt.figure(figsize=(10,7))
x1 = range(0, 15)
y1 = [m.sqrt(n) for n in x1]
plt.plot(x1, y1, c = 'dodgerblue', label = '$\sqrt{x}$')
plt.title('square root of x')

x2 = range(-5, 5)
y2 = [n ** 3 for n in x2]
plt.plot(x2, y2, c = 'seagreen',label = '$x^3$')
plt.title('x^3')

x3 = range(-5, 5)
y3 = [2 ** n for n in x3]
plt.plot(x3, y3, c = 'orangered', label = '$2^x$')
plt.title('2^x')

x4 = range(0, 15)
y4 = [1/(n + 1) for n in x4]
plt.plot(x4, y4, c = 'indigo', label = '$1/(x+1)$')

plt.xlim(0, 10)
plt.ylim(0, 10)

plt.title('Four Functions')
plt.legend()
plt.show()
```
